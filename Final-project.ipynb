## 2.  Understanding the data
This report is based on the **Immunisation dataset** which contains the estimates of routine immunisation, across 195 countries over the years 1990 to 2017. The dataset contains the figures from 1980 onwards. But since the values are quite sparse across the countries between the years 1980 to 1989, thus the data is analysed from 1990 onwards. These 195 countries are divided into nine regions: East Asia & Pacific, Eastern & Southern Africa, Eastern Europe & Central Asia, Latin America & Caribbean, Middle East & North Africa, North America, South Asia, West & Central Africa, and Western Europe.

The data is recorded for the following vaccines:
1. <i>DTP1</i>: This is the first dose of DTP vaccine. The number is recorded as percentange of surviving infants who received the first dose of DTP.
2. <i>DTP3</i>: This is the third dose of DTP vaccine. The number is recorded as percentange of surviving infants who received the third dose of DTP.
3. <i>HepBB</i>: This is the first dose of Hepatitus B vaccine. The number is recorded as percentange of live births who received hepatitus B vaccine within 24 hours of birth.
4. <i>HebB3</i>: This is the third dose of Hepatitus B vaccine. The number is recorded as percentange of surviving infants who received third dose of hepatitus B vaccine.
5. <i>Hib3</i>: This is the third dose of haemophilus influenzae type B (Hib) vaccine. The number is recorded as percentange of surviving infants who received third dose of Hib vaccine.
6. <i>MCV1</i>: This is the first dose of measles vaccine. The number is recorded as percentange of surviving infants who received first dose of measles vaccine.
7. <i>MCV2</i>: This is the second dose of measles vaccine. The number is recorded as percentange of surviving infants who received second dose of measles vaccine.

The findings from the immunisation dataset is then combined with **Deaths dataset** which contains the number of deaths recorded across 195 countries over the years 1990 to 2016. These deaths are caused by vaccine preventable diseases like, diptheria, tetanus, haemophilus influenzae type B, hepatitus and measles.

The next dataset introduced is **Vaccine Expenditure dataset**. This dataset consists of the amount of money (in US dollars) spent on vaccination by 195 countries (from all sources) from the year 2006 to 2017.

Another dataset which is analysed is based on the **GDP growth**. This dataset contains GDP growth rates in percentage from the year 1960 to 2017. 

All the above datasets are downloaded from different sources. The trends from all the datasets are combined to find how all these figures are related to each other. Since the time dimension varies across the four datasets, so the data is analysed for a common time period.

## 4.  The Immunisation dataset

The main focus of our analysis is the <i>Immunisation dataset</i> which contains vaccine coverage estimates from UNICEF and WHO. The next step is to import this dataset into dataframes, which are further used to carry out the analysis. The dataframes are cleaned and the null values are replaced with zeros.

```python
# Reading and cleaning data 

# Reading DTP1 from excel file to dataframe
DTP1_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='DTP1', usecols='A:C,E:AP')
DTP1_data = DTP1_data.fillna(0)

# Reading DTP3 from excel file to dataframe
DTP3_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='DTP3', usecols='A:C,E:AP')
DTP3_data = DTP3_data.fillna(0)

# Reading HepBB from excel file to dataframe
HepBB_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='HepBB', usecols='A:C,E:AP')
HepBB_data = HepBB_data.fillna(0)

# Reading HepB3 from excel file to dataframe
HepB3_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='HepB3', usecols='A:C,E:AP')
HepB3_data = HepB3_data.fillna(0)

# Reading MCV1 from excel file to dataframe
MCV1_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='MCV1', usecols='A:C,E:AP')
MCV1_data = MCV1_data.fillna(0)

# Reading MCV2 from excel file to dataframe
MCV2_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='MCV2', usecols='A:C,E:AP')
MCV2_data = MCV2_data.fillna(0)

# Reading Hib3 from excel file to dataframe
Hib3_data = pd.read_excel('ImmunisationUNICEF.xlsx', sheet_name='Hib3', usecols='A:C,E:AP')
Hib3_data = Hib3_data.fillna(0)
```

### 4.1 The Basics
Let's start our analysis by plotting a simple line graph. The dataset contains the figures from 1980 onwards. But since the values are quite sparse across the countries between the years 1980 to 1989, thus the data is analysed from 1990 onwards. The line graph shows the average immunisation rates for different vaccines across 195 countries over the years 1990 to 2016. The calculations are done on year-wise basis. The average rate for each vaccine is calculated across 195 countries for each year. 

Let's take an example to see how the calculations are done. For the year 1990, the average rate of <i>DTP1 vaccine</i> (i.e., first dose of DTP vaccine) across 195 countries is calculated by taking the mean of vaccination rate across 195 countries. This value is plotted as the average vaccination rate for first dose of DTP vaccine against the year 1990. In the same way, the average vaccine rate for first dose of DTP is calculated for the years 1991, 1992, 1993, and so on. 
Similarly, all the average vaccination values are calculated for each of the seven vaccines year-wise from the year 1990 till 2016.

```python
# Plotting line graph

year = list(range(1990,2018))
Y_DTP1, Y_DTP3, Y_HepBB, Y_HepB3, Y_MCV1, Y_MCV2, Y_Hib3 = [], [], [], [], [], [], []
Y_avg_immunisation = []

# Calculating average immunisation rates for different vaccines across 195 countries over the years 1990 to 2016
for i in year:
    Y_DTP1 += [DTP1_data[str(i)].mean()]
    Y_DTP3 += [DTP3_data[str(i)].mean()]
    Y_HepBB += [HepBB_data[str(i)].mean()]
    Y_HepB3 += [HepB3_data[str(i)].mean()]
    Y_MCV1 += [MCV1_data[str(i)].mean()]
    Y_MCV2 += [MCV2_data[str(i)].mean()]
    Y_Hib3 += [Hib3_data[str(i)].mean()]

# Calculating the average immunisation rate across 195 countries over the years 1990 to 2016
for i in range(0,27):
    Y_avg_immunisation += [(Y_DTP1[i] + Y_DTP3[i] + Y_HepBB[i] + Y_HepB3[i] + Y_MCV1[i] + Y_MCV2[i] + Y_Hib3[i])/7]

# Plotting line graph of average immunisation rates for different vaccines    
plt.figure(figsize=(15,7))
trace1 = go.Scatter(x = year, y = Y_DTP1, mode = 'lines+markers', name = 'DTP (First dose)')
trace2 = go.Scatter(x = year, y = Y_DTP3, mode = 'lines+markers', name = 'DTP (Third dose)')
trace3 = go.Scatter(x = year, y = Y_HepBB, mode = 'lines+markers', name = 'Hepatitus (First dose)')
trace4 = go.Scatter(x = year, y = Y_HepB3, mode = 'lines+markers', name = 'Hepatitus (Third dose)')
trace5 = go.Scatter(x = year, y = Y_MCV1, mode = 'lines+markers', name = 'Measles (First dose)')
trace6 = go.Scatter(x = year, y = Y_MCV2, mode = 'lines+markers', name = 'Measles (Second dose)')
trace7 = go.Scatter(x = year, y = Y_Hib3, mode = 'lines+markers', name = 'Hib (Third dose)')

data = [trace1, trace2, trace3, trace4, trace5, trace6, trace7]
layout=go.Layout(title='Average immunisation rates for different vaccines worldwide (across 195 countries) over the years 1990 to 2017', xaxis={'title':'Year'}, yaxis={'title':'Percentage'})
figure=go.Figure(data=data,layout=layout)
iplot(figure, filename='line-mode')
```

The above line graph is an interactive chart. It shows the average immunisation rates for different vaccines for a 28 years period.
